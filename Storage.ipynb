```python
# Import essential packages
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
from bs4 import BeautifulSoup
```

```python
# Class for storage. Or maybe use pandas dataframe
class StorageUnit:
    """
    Define a class to store related info of one product
    :param name
    """
    def __init__(self,name:str="none",html_link:str="none",price:str="0.0 euro",rating:str="none"):
        self.name = name
        self.html_link= html_link
        self.price = price
        self.rating = rating
    def __str__(self):
        return "name:\t"+ self.name + "\nlink:\t" + self.html_link +"\nprice:\t"+ str(self.price)+"\nrating\t:"+self.rating
```

```python
# Get html source from certain web page
def load_page(page_num:int):
    driver.get(url=base_url+"?page="+str(base_page))
    sleep(5)
    soup_obj=BeautifulSoup(driver.page_source,'lxml')
    return soup_obj
```

```python
# Detect if there's a next page
def go_next(soup_obj:BeautifulSoup):
    next_page_disabled = soup_obj.find(name='a',class_='glyphicon glyphicon-menu-right').attrs.get('disabled')
    if next_page_disabled=='false':
        return True
    else:
        return False
```

```python
# Analyze html source
def process(soup_obj:BeautifulSoup,store:list):
    items = soup_obj.find_all(name='div',attrs={'class':'caption'})
    for item in items:
        if item.find(class_='product-rating') is not None:
            rating=item.find(class_='product-rating').get_text()
        else: rating="Not Specified"
        store.append(StorageUnit(name=item.h3.get_text(),html_link=item.h3.a.attrs['href'],price=item.find(class_="price").get_text(),rating=rating))

```

```python
# Initial Setup
base_url="https://hobbii.fr/fils"
base_page=1
driver = webdriver.Chrome(executable_path="/Users/luciezhang/Téléchargements/chromedriver_mac64/chromedriver.exe")
storage = list[StorageUnit]()
```

```python
# Loop through available pages and store results
while True:
    soup=load_page(base_page)
    process(soup,storage)
    print("page number "+str(base_page)+" has finished\n")
    base_page=base_page+1

    if not go_next(soup):
        break
```
